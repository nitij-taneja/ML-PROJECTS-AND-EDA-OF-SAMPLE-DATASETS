```python
import pandas as pd
```

```python
import sweetviz
df = pd.read_csv("COVID clinical trials.csv")
```

```python
df.head()
```

```python
df.info
```

```python
df.shape
```

```python
from datetime import datetime
```

```python
st=pd.Series(df['Start Date'])
res=pd.to_datetime(st)
df['Start']=res.dt.strftime("%Y-%m-%d")

st1=pd.Series(df['Completion Date'])
res1=pd.to_datetime(st1)
df['completion']=res1.dt.strftime("%Y-%m-%d")


df.head()
```

```python
df['Start Month']= pd.DatetimeIndex(df['Start']).month
df['Start Year']= pd.DatetimeIndex(df['Start']).year

df['Completion Month']= pd.DatetimeIndex(df['completion']).month
df['Completion  Year']= pd.DatetimeIndex(df['completion']).year
```

```python
my_report = sweetviz.analyze([df, "Covid 19"],target_feat='Rank')
```

```python
my_report.show_html('Report_covid19_critical_trial_eda.html')
```
